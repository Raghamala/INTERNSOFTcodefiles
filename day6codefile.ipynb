```python
import time
import numpy as np
M=25
N=32
x=np.linspace(0.0,1.0,M,endpoint=False)
y=np.linspace(0.0,1.0,N,endpoint=False)
```

```python
t0=time.time()
g=np.zeros((M,N))
for i in range(M):
  for j in range(N):
    g[i,j]=np.sin(2.0*np.pi*x[i]*y[j])
    print(g)
t1=time.time()
print("time elapsed:{}s".format(t1-t0))
```

```python
x2d,y2d=np.meshgrid(x,y,indexing="ij")
print(x2d[:,0])
print(x2d[0,:])
print(y2d[:,0])
print(y2d[0,:])
```
